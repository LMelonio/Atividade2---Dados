```python
#Comandos para importar bibliotecas
import numpy as np
import pandas as pd
```

```python
#comando para importar os dados
df=pd.read_csv('/content/agricultural_raw_material.csv')
```

```python
#Mostrar os dados
df.info
#encontrar os valores nulos
df.isnull().sum()
```

```python
#Comando para visualizar as 5 primeiras linhas
df.head()
```

```python
# Alterando %, "," and "-" para visualizar o dados com mais facilidade.
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)
```

```python
# Eliminando valores nulos
df.dropna(inplace=True)
# Verificando se a alteração foi feita
df.isnull().sum()
```

```python
# alterando os campos para float
lst = ["Coarse wool Price", "Coarse wool price % Change", "Copra Price", "Copra price % Change", "Cotton price % Change","Fine wool Price", "Fine wool price % Change", "Hard log price % Change", "Hard sawnwood price % Change", "Hide price % change", "Plywood price % Change", "Rubber price % Change", "Softlog price % Change", "Soft sawnwood price % Change", "Wood pulp price % Change"]
df[lst] = df[lst].astype("float")
df.dtypes
```

```python
#Mostrando as linhas mais recentes
df.tail()
```

```python
#Bibliotecas para exploração
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
#Este comando define uma lista contém os nomes das colunas dos dados que serão utilizados para calcular a matriz de correlação.
raw_data=['Coarse wool Price',  'Copra Price','Cotton Price', 'Fine wool Price',  'Hard log Price', 'Hard sawnwood Price',
 'Hide Price', 'Plywood Price', 'Rubber Price', 'Softlog Price', 'Soft sawnwood Price', 'Wood pulp Price']
#comando para calcular a matriz de correlação para as coluna "df" e "raw_data".
corrmat = df[raw_data].corr()
#Criando um gráfico com um tamanho de 12 polegadas de largura por 9 polegadas de altura.
fig = plt.figure(figsize = (12, 9))
#criando uma máscara booleana para ocultar a metade superior da matriz.
mask = np.triu(np.ones_like(corrmat, dtype=bool))
#Criando mapa de calor utilizando a biblioteca seaborn.
sns.heatmap(corrmat, vmax = .8,mask=mask, square = True, annot = True)
plt.show()
```

Este código é útil para visualizar a correlação entre diferentes variáveis em um conjunto de dados, o que pode ajudar na identificação de padrões e relacionamentos entre as variáveis.

```python
#Este comando define uma lista "changelist" que contém os nomes das colunas do "df" que representam mudanças percentuais de preços.
changelist=['Copra price % Change','Softlog price % Change','Rubber price % Change','Cotton price % Change','Coarse wool price % Change','Fine wool price % Change','Hard log price % Change','Hard sawnwood price % Change','Hide price % change','Plywood price % Change','Soft sawnwood price % Change','Wood pulp price % Change']
#Este comando inicia um loop sobre os elementos da lista "changelist".
for i in range(len(changelist)):
    plt.figure(figsize=(12,12))
    df[changelist[i]].hist(figsize=(11, 9), linewidth=1)
    plt.xlabel('% Change')
    plt.ylabel('count')
    plt.legend(changelist[i:],loc='upper center',bbox_to_anchor=(1.2,1))
```

O objetivo deste código é plotar histogramas para visualizar a distribuição das mudanças percentuais de preços ao longo do tempo para diferentes produtos. O loop permite que todos os histogramas sejam plotados em uma única figura, facilitando a comparação entre as distribuições de diferentes produtos.

```python
#criando um gráfico
plt.figure(figsize=(10, 10))
# Criando um gráfico de linhas para as colunas "Cotton Price" e "Rubber Price" . Ele exibirá a tendência desses preços ao longo do tempo.
plt.plot(df[['Cotton Price','Rubber Price']])
#adicionando um título ao gráfico.
plt.title("Raw-Materials price comparision")
#Este comando adiciona um rótulo ao eixo x do gráfico, indicando que os valores representam os anos.
plt.xlabel('Years')
#Este comando adiciona um rótulo ao eixo y do gráfico, indicando que os valores representam os preços das matérias-primas.
plt.ylabel('Prices')
#adicionando umaa legenda ao gráfico, especificando os nomes das séries de dados. A localização da legenda é definida para o canto superior direito, fora da área do gráfico, para evitar obstrução dos dados.
plt.legend(['Cotton Price','Rubber Price'],loc='upper center',bbox_to_anchor=(1.2,1))
```

O objetivo deste código é visualizar a tendência dos preços de algodão e borracha ao longo dos anos em um único gráfico de linhas, facilitando a comparação entre os dois.
**Resultado:**
- O **algodão** possui o menor preço.

```python
# Criando uma gráfico
plt.figure(figsize=(10, 10))
# Plota um gráfico de linhas para as colunas que representam os preços das matérias-primas.
plt.plot(df[['Copra Price','Softlog Price','Rubber Price','Cotton Price','Coarse wool Price','Fine wool Price','Hard log Price','Hard sawnwood Price','Hide Price','Plywood Price','Soft sawnwood Price','Wood pulp Price']])
# Adiciona um título ao gráfico
plt.title("Raw-Materials price comparision")
# Adiciona um rótulo ao eixo x do gráfico
plt.xlabel('Years')
# Adiciona um rótulo ao eixo y do gráfico
plt.ylabel('Prices')
plt.legend(['Copra Price','Softlog Price','Rubber Price','Cotton Price','Coarse wool Price','Fine wool Price','Hard log Price','Hard sawnwood Price','Hide Price','Plywood Price','Soft sawnwood Price','Wood pulp Price'],loc='upper center',bbox_to_anchor=(1.2,1))
```

Este código Python está plotando um gráfico de linhas comparando os preços de várias matérias-primas ao longo dos anos. Cada matéria-prima é representada por uma linha no gráfico, permitindo a visualização da tendência dos preços ao longo do tempo e facilitando a comparação entre elas.
